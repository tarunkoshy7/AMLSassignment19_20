```python
import lab2_landmarks as l2
import landmarks_test as ltest
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, accuracy_score, log_loss
from sklearn.model_selection import train_test_split
from sklearn import svm
```

We have to extract the features from each of the images and split them into training, validation, and test sets. Feature extraction is done using dlib, which I called from the lab2_landmarks file given in lab 2 of this course. Once the features had been extracted, I used scikit-learn to split the dataset into a 60-20-20 train-val-test split.

```python
def get_data():
    """
    Get the extracted dlib features from the lab2_landmarks file and split into train, validation, and test sets
    """
    X, y = l2.extract_features_labels()

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
    X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.25, random_state=1)

    X_train_final = X_train.reshape(X_train.shape[0],X_train.shape[1]*X_train.shape[2])
    X_val_final = X_val.reshape(X_val.shape[0],X_val.shape[1]*X_val.shape[2])
    X_test_final = X_test.reshape(X_test.shape[0],X_test.shape[1]*X_test.shape[2])

    return X_train_final, y_train, X_val_final, y_val, X_test_final, y_test
```

```python
X_train, y_train, X_val, y_val, X_test, y_test = get_data()
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_val.shape)
print(y_val.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
def trainSVM(training_images, training_labels, test_images, test_labels, lambd, degreeInput=3, kernelInput='rbf', gammaInput='auto'):
    """
    Train the classifier using the appropriate hyperparameters, fit it to the training labels, 
    and output the predictions
    """
    classifier = svm.SVC(C=lambd, kernel=kernelInput, degree=degreeInput, gamma=gammaInput)
    classifier.fit(training_images, training_labels)
    pred = classifier.predict(test_images)
    
    return pred
```

```python
def crossValidation(X_train, y_train, X_val, y_val, degree=3, kernel='rbf', gamma='auto'):
    """
    Manually train the classifier using different C values, plot the graph of Error vs C to observe
    the differences and return the C with the lowest error
    """
    lambda_vec = np.array([0.001, 0.01, 0.1, 1, 10])
    
    error_train = np.zeros(len(lambda_vec))
    error_val = np.zeros(len(lambda_vec))
    
    for i in range(len(lambda_vec)):
        lambd = lambda_vec[i]
        y_train_pred = trainSVM(X_train, y_train, X_train, y_train, lambd, degreeInput=degree, kernelInput=kernel, gammaInput=gamma)
        y_val_pred = trainSVM(X_train, y_train, X_val, y_val, lambd, degreeInput=degree, kernelInput=kernel, gammaInput=gamma)
        
        error_train[i] = log_loss(y_train, y_train_pred)
        error_val[i] = log_loss(y_val, y_val_pred)

    print('Training Error:', error_train)
    print('Validation Error:', error_val)

    minErrorIndex = np.argmin(error_val)
    minError = np.amin(error_val)
    print('Min Validation Error Value:', minError)
    print('Ideal C Value:', lambda_vec[minErrorIndex])
    
    plt.plot(lambda_vec, error_train, label='Train')
    plt.plot(lambda_vec, error_val, label='Cross Validation')
    plt.legend()
    plt.xlabel('C')
    plt.ylabel('Error')
    plt.title('Validation Curve')
    plt.show()    
        
    return lambda_vec[minErrorIndex], minError
```

```python
def learningCurve(X_train, y_train, X_val, y_val, lambd, degree=3, kernel='rbf', gamma='auto'):
    """
    Plot the learning curve of the training and validation set errors over number of training examples
    """
        m = X_train.shape[0]
        error_train = np.zeros(int(m/10))
        error_val = np.zeros(int(m/10))
        
        for i in range(10, m, 10):
            y_train_pred = trainSVM(X_train[0:i], y_train[0:i], X_train[0:i], y_train[0:i], lambd, degreeInput=degree, kernelInput=kernel, gammaInput=gamma)
            y_val_pred = trainSVM(X_train[0:i], y_train[0:i], X_val, y_val, lambd, degreeInput=degree, kernelInput=kernel, gammaInput=gamma)
            
            error_train[int((i-10)/10)] = log_loss(y_train[0:i], y_train_pred)
            error_val[int((i-10)/10)] = log_loss(y_val, y_val_pred)
            
        plt.plot(np.arange(10, m, 10), error_train, label='Train')
        plt.plot(np.arange(10, m, 10), error_val, label='Cross Validation')
        plt.legend()
        plt.xlabel('Number of Training Examples')
        plt.ylabel('Error')
        plt.title('Learning Curve')
        plt.show()
        
        return None
```

# Linear Test

```python
lambdaLinear = crossValidation(X_train, y_train, X_val, y_val,kernel='linear')[0]
```

```python
learningCurve(X_train, y_train, X_val, y_val, lambdaLinear, kernel='linear')
```

```python
pred_train=trainSVM(X_train, y_train, X_train, y_train, lambdaLinear, kernelInput='linear')
print('Accuracy:', accuracy_score(y_train, pred_train))
print(classification_report(y_train, pred_train))
```

```python
pred_val=trainSVM(X_train, y_train, X_val, y_val, lambdaLinear, kernelInput='linear')
print('Accuracy:', accuracy_score(y_val, pred_val))
print(classification_report(y_val, pred_val))
```

```python
%time pred_test_linear=trainSVM(X_train, y_train, X_test, y_test, lambdaLinear, kernelInput='linear')
print('Accuracy:', accuracy_score(y_test, pred_test_linear))
print(classification_report(y_test, pred_test_linear))
```

# RBF Test

```python
def gammaCheck(X_train, y_train, X_val, y_val, kernel='rbf'):
    gamma_vec = np.array([0.0001, 0.0003, 0.001, 0.003, 0.01, 0.03, 0.1, 0.3, 1])
    
    lambdaGamma = np.zeros(len(gamma_vec))
    minError = np.zeros(len(gamma_vec))
    
    for i in range(len(gamma_vec)):
        gamma = gamma_vec[i]
        lambdaGamma[i], minError[i] = crossValidation(X_train, y_train, X_val, y_val, kernel=kernel, gamma=gamma)

    print('Ideal Lambdas:', lambdaGamma)
    print('Minimum Validation Errors:', minError)

    minminErrorIndex = np.argmin(minError)
    idealLambda = lambdaGamma[minminErrorIndex]
    print('Ideal Gamma:', gamma_vec[minminErrorIndex])
    print('Ideal Degree Lambda:', idealLambda)
    
    plt.plot(gamma_vec, minError, label='Gamma')
    plt.xlabel('Gamma')
    plt.ylabel('Minimum Error')
    plt.title('Gamma Validation Curve')
    plt.show()    
        
    return gamma_vec[minminErrorIndex], idealLambda
```

```python
idealGamma, lambdaRBF = gammaCheck(X_train, y_train, X_val, y_val)
```

```python
learningCurve(X_train, y_train, X_val, y_val, lambdaRBF, gamma=idealGamma)
```

```python
pred_test_rbf=trainSVM(X_train, y_train, X_test, y_test, lambdaRBF, gammaInput=idealGamma)
print('Accuracy:', accuracy_score(y_test, pred_test_rbf))
print(classification_report(y_test, pred_test_rbf))
```

```python
%time pred_test_new_RBF=trainSVM(X_train, y_train, X_test_new, y_test_new, lambdaRBF, gammaInput=idealGamma)
print('New Test Accuracy:', accuracy_score(y_test_new, pred_test_new_RBF))
print(classification_report(y_test_new, pred_test_new_RBF))
```

# Polynomial Test

```python
def degreeCheck(X_train, y_train, X_val, y_val, kernel='poly'):
    degree_vec = np.array([1, 2])
    
    lambdaPoly = np.zeros(len(degree_vec))
    minError = np.zeros(len(degree_vec))
    
    for i in range(len(degree_vec)):
        degree = degree_vec[i]
        lambdaPoly[i], minError[i] = crossValidation(X_train, y_train, X_val, y_val, degree=degree, kernel=kernel)

    print('Ideal Lambdas:', lambdaPoly)
    print('Minimum Validation Errors:', minError)

    minminErrorIndex = np.argmin(minError)
    idealLambda = lambdaPoly[minminErrorIndex]
    print('Ideal Degree:', degree_vec[minminErrorIndex])
    print('Ideal Degree Lambda:', idealLambda)
    
    plt.plot(degree_vec, minError, label='Degree')
    plt.xlabel('Degree')
    plt.ylabel('Minimum Error')
    plt.title('Polynomial Validation Curve')
    plt.show()    
        
    return degree_vec[minminErrorIndex], idealLambda
```

```python
polyDegree, lambdaPoly = degreeCheck(X_train, y_train, X_val, y_val, kernel='poly')
```

```python
learningCurve(X_train, y_train, X_val, y_val, lambdaPoly, degree=polyDegree, kernel='poly')
```

```python
pred_test_poly=trainSVM(X_train, y_train, X_test, y_test, lambdaPoly, degreeInput=polyDegree, kernelInput='poly')
print('Accuracy:', accuracy_score(y_test, pred_test_poly))
print(classification_report(y_test, pred_test_poly))
```
